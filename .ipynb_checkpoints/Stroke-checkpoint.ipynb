## - Dropping Null Values

```python
df = df.dropna()
df = df.reset_index()
df.isnull().sum()
```

```python
ordinal_features = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
class_category_order = [
['Male', 'Female', 'Other'], 
['Yes', 'No'],
['Private', 'Self-employed', 'Govt_job', 'children', 'Never_worked'],
['Urban', 'Rural'],
['formerly smoked', 'never smoked', 'smokes', 'Unknown']
]
encoder = OrdinalEncoder(categories = class_category_order)
df[ordinal_features] = encoder.fit_transform(df[ordinal_features])
df
```

# Outliers

```python
def CountOutliers(series):
    series = pd.to_numeric(series)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((series < lower_bound) | (series > upper_bound)).sum()

numerical_features = ['age', 'avg_glucose_level', 'bmi']

for feature in numerical_features:
    print(f"{feature} : {CountOutliers(df[feature])}")
```

## - With IQR Filtering

```python
Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['bmi'] > lower) & (df['bmi'] < upper)]
```

```python
for feature in numerical_features:
    plt.figure()
    plt.boxplot(df[feature])
    plt.title(feature)
```

# Modeling

## - Soft Margin RBF Kernel

```python
x = df.drop('stroke', axis = 1)
y = df['stroke']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
svm = SVC(kernel='rbf', C=1.0, gamma='scale')
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

## PCA

#### - Linear Kernel

```python
numerical_features = ['age', 'avg_glucose_level', 'bmi']
scaler = StandardScaler()

for feature in numerical_features:
    df[feature] = scaler.fit_transform(df[[feature]])

x = df.drop('stroke', axis = 1)
y = df['stroke']

pca = PCA(n_components=2)
x = pca.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
svm = SVC(kernel='linear', C=1.0, gamma='scale')
svm.fit(x_train, y_train)
x_min, x_max = x_train[:, 0].min() - 1, x_train[:, 0].max() + 1
y_min, y_max = x_train[:, 1].min() - 1, x_train[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)

Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap=plt.cm.Paired, edgecolors='k')
plt.title('SVM Decision Boundary')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.show()

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
